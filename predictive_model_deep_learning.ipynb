**TERCER PARTE:**

En este sector se aplicarГЎn los modelos de Machine Learning seleccionados:
 - El modelo que serГЎ utilizado como un baseline es RegresiГіn LogГӯstica Binaria
 - Luego se aplicarГЎ Bayes Ingenuo, ya que selecciona la clase a la que va a pertenecer la observaciГіn, calculando una probabilidad de pertenencia a la misma. AdemГЎs, a pesar de que asume que todas las variables son independientes se maneja muy bien con variables categoricas
 - Por Гәltimo se aplicarГЎ Boosting con AdaBoost (es el mГЎs optimo para variables binarias), donde se centra en el error de su iteraciГіn en vez de lo que realiza bien, donde aprenderГЎ de los errores y permitirГЎ un mejor entrenamiento. TambiГ©n lo compararГ© con XGBoost, ya que a pesar de que se maneja mejor con variables multinomiales probarГ© ver si termina siendo mГЎs Гіptimo en este caso (Гәnicamente es para probar si cambia en algo o logra una mejora).

 PD: PensГ© en aplicar Clasificador de Soporte Vectorial, debido a que separa los datos por un hiperplano, creando una frontera de decisiГіn lineal o no lineal en un espacio de mayor dimensiГіn.

```python
# Corro la regresiГіn logГӯstica
clf = LogisticRegression(random_state=0, max_iter=10000).fit(x_train, y_train)

# Miro los coeficientes que se obtuvieron
print(clf.coef_)

# Miro el intercept
print(clf.intercept_)

plt.figure(figsize=(10,6))
plt.bar(x.columns, clf.coef_[0])
plt.xticks(rotation=45)
plt.title("Importancia de cada variable en la regresiГіn logГӯstica")
plt.ylabel("Coeficiente")
plt.grid(True)
plt.show()

# Calculo el accuracy en train y test
print(clf.score(x_train, y_train))
print(clf.score(x_test, y_test))
```

```python
# Crear una tabla con los coeficientes y su significancia estadГӯstica


# Obtener los coeficientes del modelo
coeficientes = clf.coef_[0]
nombres_features = x.columns

# Crear un DataFrame con los coeficientes
df_coef = pd.DataFrame({
  'Feature': nombres_features,
  'Coeficiente': coeficientes,
  'Abs_Coeficiente': np.abs(coeficientes)
})

# Ordenar por valor absoluto del coeficiente (descendente)
df_coef = df_coef.sort_values('Abs_Coeficiente', ascending=False)

# Calcular el error estГЎndar y el valor p para cada coeficiente
# Usamos la matriz de covarianza del modelo logГӯstico

# Predicciones en el conjunto de entrenamiento
y_pred_proba = clf.predict_proba(x_train)

# Calcular la matriz de informaciГіn de Fisher (Hessian)
# Para regresiГіn logГӯstica: H = X^T * W * X, donde W es la matriz diagonal de pesos
W = np.diag(y_pred_proba[:, 0] * y_pred_proba[:, 1])  # w_i = p_i * (1 - p_i)
X_train_array = x_train.values.astype(float)
hessian = X_train_array.T @ W @ X_train_array

# Matriz de covarianza = inversa del Hessian
try:
  cov_matrix = np.linalg.inv(hessian)
  std_errors = np.sqrt(np.diag(cov_matrix))
except:
  # Si la matriz no es invertible, usar una aproximaciГіn
  std_errors = np.ones(len(coeficientes)) * 0.1

# Calcular z-scores y p-values
z_scores = coeficientes / std_errors
p_values = 2 * (1 - stats.norm.cdf(np.abs(z_scores)))

# AГұadir al DataFrame
df_coef['Std_Error'] = std_errors
df_coef['Z_Score'] = z_scores
df_coef['P_Value'] = p_values
df_coef['Significancia'] = df_coef['P_Value'].apply(
  lambda p: '***' if p < 0.001 else ('**' if p < 0.01 else ('*' if p < 0.05 else 'No significativo'))
)

# Mostrar el top 15 caracterГӯsticas
top_features = df_coef.head(15)[['Feature', 'Coeficiente', 'Std_Error', 'Z_Score', 'P_Value', 'Significancia']]

print("=" * 120)
print("TOP 15 COVARIABLES CON MAYOR COEFICIENTE (EN VALOR ABSOLUTO)")
print("=" * 120)
print(top_features.to_string(index=False))
print("\n*** p < 0.001 | ** p < 0.01 | * p < 0.05")
```

**PRГ“XIMO PASO:** Aplicar Redes Neuronales y ver si hay mejora

Aplicamos unicamente la red neuronal con Early Stopping para que determine hasta donde hay mejora y no corra indeterminadamente sin sentido.
Al ver que es crucial escalar los datos para aplicar redes neuronales, posterior a probarlo sin ello, se lo voy a aplicar para ver si realmente representa una mejora.

En este caso no aplico Dropout, ni patience, o demГЎs cosas ya que quiero probarlos por separado y ver si mejoran en algo el modelo y cual es el verdadero problema.

```python
# Definir el modelo de red neuronal
model = Sequential([
    # Capa de entrada y primera capa oculta con activaciГіn ReLU y regularizaciГіn L2
    Dense(32, activation='relu', input_shape=(x_train_dum.shape[1],), kernel_regularizer=l2(0.001)),

    # Segunda capa oculta con activaciГіn ReLU y regularizaciГіn L2
    Dense(64, activation='relu', kernel_regularizer=l2(0.001)),

    # Tercera capa oculta con activaciГіn ReLU y regularizaciГіn L2
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),

    # Capa de salida con activaciГіn sigmoide para clasificaciГіn binaria
    Dense(1, activation='sigmoid')
])

# Compilar el modelo
# Usamos 'Adam' como optimizador, 'binary_crossentropy' como funciГіn de pГ©rdida para clasificaciГіn binaria,
# y 'accuracy' como mГ©trica principal.
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Mostrar un resumen del modelo
print("\nResumen del modelo de red neuronal:")
model.summary()
```

```python
# Definir la llamada de Early Stopping
# Monitorea la 'val_loss' (pГ©rdida en el conjunto de validaciГіn)
# 'patience=10' significa que esperarГЎ 10 Г©pocas sin mejora antes de detenerse
# 'restore_best_weights=True' restaurarГЎ los pesos del modelo de la mejor Г©poca
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Entrenar el modelo
print("\nEntrenando el modelo...")
history = model.fit(
    x_train_dum, y_train_dum,
    epochs=100, # NГәmero mГЎximo de Г©pocas
    batch_size=32, # TamaГұo del lote para el entrenamiento
    validation_data=(x_val_dum, y_val_dum), # Datos de validaciГіn
    callbacks=[early_stopping], # Aplicar Early Stopping
    verbose=1 # Mostrar progreso del entrenamiento
)

print("\nEntrenamiento completado.")
```

```python
# Evaluar el modelo en el conjunto de prueba
loss, accuracy = model.evaluate(x_test_dumm, y_test_dumm, verbose=0)
print(f"\nPrecisiГіn del modelo en el conjunto de prueba: {accuracy:.4f}")
print(f"PГ©rdida del modelo en el conjunto de prueba: {loss:.4f}")

# Realizar predicciones en el conjunto de prueba
y_pred_proba = model.predict(x_test_dumm)
y_pred = (y_pred_proba > 0.5).astype(int) # Convertir probabilidades a clases binarias (0 o 1)

# Mostrar el reporte de clasificaciГіn
print("\nReporte de clasificaciГіn en el conjunto de prueba:")
print(classification_report(y_test_dumm, y_pred))
```

En este caso representa un aprendizaje bastante parecido a lo que estaba en la regresiГіn logГӯstica, ya que el modelo se vuelve experto prediciendo a la clase mayoritaria "0"(<=50K), donde su f1-score es altisimo al igual que el recall, y... si este dataset solo se basara en la variables respuesta 0 serГӯa excelente.
Pero el verdadero problema se encuentra al intentar predecir la variable 1 (>50K) que son la clase minoritaria

```python
# 1. Creamos un Callback personalizado para evaluar Test en cada Г©poca
class TestF1History(Callback):
    def __init__(self, x_train, y_train, x_val, y_val, x_test, y_test):
        self.x_train = x_train
        self.y_train = y_train
        self.x_val = x_val
        self.y_val = y_val
        self.x_test = x_test
        self.y_test = y_test
        
        # Listas para guardar la historia
        self.train_f1 = []
        self.val_f1 = []
        self.test_f1 = []

    def on_epoch_end(self, epoch, logs=None):
        # Predicciones de la Г©poca actual (umbral 0.5)
        preds_train = (self.model.predict(self.x_train, verbose=0) > 0.5).astype(int)
        preds_val = (self.model.predict(self.x_val, verbose=0) > 0.5).astype(int)
        preds_test = (self.model.predict(self.x_test, verbose=0) > 0.5).astype(int)
        
        # Calcular F1 aislado para la clase minoritaria (pos_label=1)
        self.train_f1.append(f1_score(self.y_train, preds_train, pos_label=1, zero_division=0))
        self.val_f1.append(f1_score(self.y_val, preds_val, pos_label=1, zero_division=0))
        self.test_f1.append(f1_score(self.y_test, preds_test, pos_label=1, zero_division=0))

# 2. Instanciamos nuestro callback con tus conjuntos escalados
f1_history = TestF1History(x_train_scaled, y_train_dum, x_val_scaled, y_val_dum, x_test_scaled, y_test_dumm)

# 3. Agregamos el callback al entrenamiento (mantenemos early_stopping)
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

print("\nEntrenando el modelo y calculando mГ©tricas por Г©poca...")
history = model.fit(
    x_train_scaled, y_train_dum,
    epochs=100,
    batch_size=75,
    validation_data=(x_val_scaled, y_val_dum),
    callbacks=[early_stopping, f1_history], # <--- Metemos el nuevo callback acГЎ
    verbose=1
)

print("\nEntrenamiento completado.")

# 4. Generar el grГЎfico de lГӯneas con las curvas completas
epochs_range = range(1, len(f1_history.train_f1) + 1)

plt.figure(figsize=(10, 6))
plt.plot(epochs_range, f1_history.train_f1, linestyle='-', linewidth=2, color='forestgreen', label='Train F1-Score')
plt.plot(epochs_range, f1_history.val_f1, linestyle='-', linewidth=2, color='royalblue', label='Val (Dev) F1-Score')
plt.plot(epochs_range, f1_history.test_f1, linestyle='--', linewidth=2, color='crimson', label='Test F1-Score')

# Configuraciones del grГЎfico
plt.title('EvoluciГіn del Aprendizaje: F1-Score (Clase Minoritaria >50K)', fontsize=13, pad=15, weight='bold')
plt.ylabel('Rendimiento F1-Score (0 a 1)', fontsize=11)
plt.xlabel('Гүpocas de Entrenamiento', fontsize=11)
plt.ylim(0, 1)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=11, loc='lower right')

plt.tight_layout()
plt.show()
```
